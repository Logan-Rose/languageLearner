```python
import numpy as np
import pandas as pd

df = pd.read_csv('../French Sentances.csv')
for i in range(10):
  for i in range(len(df)):
    if pd.isnull(df.iloc[i-1,0]):
      print(pd.isnull(df.iloc[i-1, 0]))
      if not pd.isnull(df.iloc[i-1, 0]):
        df.iloc[i][0] = df.iloc[i-1][0] + df.iloc[i][0]
        df.drop(index=[i-1])


print('---')

df.to_csv('../tmp.csv')
```
